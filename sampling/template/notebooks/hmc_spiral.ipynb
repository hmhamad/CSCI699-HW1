```python
import matplotlib.pyplot
import numpy
```

```python
from utils.density import continuous_energy_from_image, prepare_image
img_name = 'spiral' # {'labrador','spiral'}
img = matplotlib.image.imread(f'./data/{img_name}.jpg')
crop_ranges = (10, 710, 240, 940) if img_name=='labrador' else None
density, energy = prepare_image(
    img, crop=crop_ranges, white_cutoff=225, gauss_sigma=3, background=0.01
)
fig = matplotlib.pyplot.figure(figsize=(10, 10))
ax = fig.add_subplot(1, 1, 1)
ax.imshow(energy)
matplotlib.pyplot.show()
```

```python
d = 2  # dimensionality 
num_samples = 10000
k = 40
epsilon = 0.5

trajectories = numpy.load(f'samples/samples_{img_name}_niter_{num_samples}_k_{k}_eps_{epsilon}.npy')
fig = matplotlib.pyplot.figure(figsize=(10, 10))
ax = fig.add_subplot(1, 1, 1)
ax.scatter(numpy.array(trajectories)[:, 1], numpy.array(trajectories)[:, 0], s=10)
ax.imshow(density, alpha=0.3)
matplotlib.pyplot.show()
```
